```python
import os

import pandas as pd
```

```python
old_csv = [x for x in os.listdir("data") if x.endswith('edited.csv')]
old_csv
```

```python
def actual_calc(row):
    if row['Value USD'] > row['USD 2']:
        return 1
    return 0
```

```python
old_csv_filenames = []
for filename in old_csv:
    print('data/%s'%filename)
    old_csv_filenames.append('data/%s'%filename)
```

```python
df = pd.read_csv(old_csv_filenames[0])
df.head()
```

```python
df['USD 2'] = df['Value USD'].shift(-1)
df.dropna(inplace=True)
df.head()
```

```python
df['Actual'] = df.apply(actual_calc, axis=1)
df.head()
```

```python
df.drop(['USD 2'], axis=1, inplace=True)
df.head()
```

```python
for filename in old_csv_filenames:
    df = pd.read_csv(filename)
    
    df['USD 2'] = df['Value USD'].shift(-1)
    df.dropna(inplace=True)
    df['Actual'] = df.apply(actual_calc, axis=1)
    df.drop(['USD 2'], inplace=True, axis=1)
    
    df.to_csv(filename, index=False)
```
